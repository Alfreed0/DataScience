```python
df_0 = df[df['postura'] == 0]
df_1 = df[df['postura'] == 1]
df_2 = df[df['postura'] == 2]
df_3 = df[df['postura'] == 3]
df_4 = df[df['postura'] == 4]
df_5 = df[df['postura'] == 5]
df_6 = df[df['postura'] == 6]
```

```python
df_concat = pd.concat([df_0, df_1, df_2, df_3, df_4, df_5, df_6])
```

```python
df_concat = df_concat.sample(frac=1).reset_index(drop=True)
```

```python
# Downsamplo tipo_postura using the same number of samples for each class
df_down = df_concat.groupby('postura').apply(lambda x: x.sample(3391, replace=True, random_state=32)).reset_index(drop=True)
```
