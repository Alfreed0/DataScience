```python
import pandas as pd
import numpy as np
```

```python
dfene = pd.read_csv('../Data/sales_2017_enero.csv', parse_dates=['date'])
dffeb = pd.read_csv('../Data/sales_2017_febrero.csv', parse_dates=['date'])
dfmar = pd.read_csv('../Data/sales_2017_marzo.csv', parse_dates=['date'])
```

```python
dfene
```

```python
dfene.dtypes
```

```python
dfene['date'].dt.dayofweek
```

```python
dftri = pd.concat([dfene, dffeb, dfmar])
```

```python
dftri
```

```python
dftri.reset_index(drop=True, inplace=True)
```

```python
dftri['day'] = dftri['date'].dt.dayofweek
```

```python
stores = pd.read_csv('../Data/stores.csv')
```

```python
stores
```

```python
dfm = dftri.merge(stores, on='store_nbr') #left_on="atributte" right_on = "atributte" in case they aren’t named the same or you wand to be more specific.
                                    #how="left" #or right and inner # adding to the left or right original table respectively. If there is no matching element it removes the element.
```

```python
dftri.merge(stores, how='right').shape
```

```python
stores.merge(dftri, how='left').shape
```

```python
dfr = dfm.groupby('state').agg({'unit_sales': ['sum', 'min', 'max', 'mean', 'median', 'std', 'size'], 'item_nbr': ['nunique', lambda x: x.nunique() / x.size]}).reset_index()
```

```python
dfm['trx_mean'] = dfm.groupby('city')['unit_sales'].transform('mean') #transform applies the function to each element of the group
```
